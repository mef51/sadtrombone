# Drift normalized by observing frequency vs. Burst Duration

```python
from frbrepeaters import driftlaw
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None  # default='warn'
%matplotlib qt

sources_filenames = [
    "frb121102_michilli_dmvariations.csv",
    "universal/data/CHIME_180916.J0158+65/dmvars_180916.csv",
    "universal/data/CHIME_FRB180814.J0422+73/dmvars_180814.csv"
]

centerDMs = [
    559.7, # pc/cm3
    348.82,
    189.4,
]

sources = []
for sfile in sources_filenames:
    df = pd.read_csv(sfile)
    df = df.set_index("name")
    df = driftlaw.computeModelDetails(df)
    sources.append(df)

names = [
    'FRB121102 DM Variations', 
    'FRB180916 DM Variations', 
    'FRB180814 DM Variations'
]
annotations = [
    [0, 1, 2, 3, 4, 5],
    [3],
    [2]
]
exclusions = [
    [],
    [],
    [180917.0],
]
for source, name, annotate, exclude in zip(sources, names, annotations, exclusions):
    frames = []
    labels = []
    for dm, color in zip(source.DM.unique(), ['r', 'y', 'b', 'k', 'g', 'c']):
        df = source[source.DM == dm]
        
        # burst exclusions
        df = df[df.drift_abs > 0] # exclude positive drifts, we assume they are non-physical and an artifact of over dedispersion
        df = df[df['drift error (mhz/ms)'] < 1e30]
        df = df.drop(exclude)
        
        df['color'] = color
        frames.append(df)
        labels.append(dm)
    
    driftlaw.plotDriftVsDuration(frames, labels, title=name, annotatei=annotate, logscale=False)
#dmvars_180916.DM.unique()
```
