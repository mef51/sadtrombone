# Approximate $\tau_0$ and scale the burst drift

$ \tau_0 \simeq \frac{b}{\cos\theta} $ Where $b$ is the minor axis of the gaussian fit to the burst autocorrelation.
Then, if $d\nu_{\text{obs}} / dt_{\text{obs}}$ is the burst drift, we expect 


### $\frac{d\nu_{\text{obs}}}{dt_{\text{obs}}}=\frac{\nu_{\text{obs}}^{2}}{\nu_{0}\tau_{0}}\simeq\frac{\nu_{\text{obs}}^{2}b}{\nu_{0}\cos\theta}$

```python
import numpy as np
import scipy.stats
import scipy.odr
import scipy.optimize
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib as mpl
from tqdm import tqdm
import functools
print = functools.partial(print, flush=True) # print doesn't happen til script ends so force it to flush... windows thing?
import pandas as pd

bursts121102 = pd.read_csv('../bursts.csv')
bursts121102 = bursts121102.set_index('name')

##########
bursts180916 = pd.merge(pd.read_csv('data/CHIME_180916.J0158+65/chime_180916_fit_params_ddm0.csv'),
                        pd.read_csv('data/CHIME_180916.J0158+65/chime_180916_param_errors_ddm0.csv'),
                        how='left', on='name')
bursts180916 = bursts180916.set_index('name')
bursts180916['time_res'] = 0.00098304 # s

#########
params180916 = ['data/CHIME_180916.J0158+65/chime_180916_fit_params_ddm0.csv',
               'data/CHIME_180916.J0158+65/chime_180916_fit_params_ddm-1.csv',
               'data/CHIME_180916.J0158+65/chime_180916_fit_params_ddm-2.csv',
               'data/CHIME_180916.J0158+65/chime_180916_fit_params_ddm0.5.csv',
               'data/CHIME_180916.J0158+65/chime_180916_fit_params_ddm1.csv']

paramerrs180916 = ['data/CHIME_180916.J0158+65/chime_180916_param_errors_ddm0.csv',
                   'data/CHIME_180916.J0158+65/chime_180916_param_errors_ddm-1.csv',
                   'data/CHIME_180916.J0158+65/chime_180916_param_errors_ddm-2.csv',
                   'data/CHIME_180916.J0158+65/chime_180916_param_errors_ddm0.5.csv',
                   'data/CHIME_180916.J0158+65/chime_180916_param_errors_ddm1.csv']

dataframes180916 = []
dmvars_180916 = None # concatenate all into one frame
dms180916 = [348.82, 348.82 - 1, 348.82 - 2, 348.82 + 0.5, 348.82 + 1]
for param, paramerr, dm in zip(params180916, paramerrs180916, dms180916):
    frame = pd.merge(pd.read_csv(param),
                     pd.read_csv(paramerr),
                     how='left', on='name')
    frame = frame.set_index('name')
    frame['time_res'] = 0.00098304 # s
    frame['DM']       = dm 
    dataframes180916.append(frame)
    if dmvars_180916 is None:
        dmvars_180916 = frame
    else:
        dmvars_180916 = pd.concat([dmvars_180916, frame])
dmvars_180916.to_csv('data/CHIME_180916.J0158+65/dmvars_180916.csv')  

#########

#########
dms180814 = [189.0 , 189.8 , 190.0, 188.9 , 189.4, 188.8, 188.7]
dm_idx = 3
bursts180814 = pd.merge(pd.read_csv('data/CHIME_FRB180814.J0422+73/chime_180814_fit_params_dm{}.csv'.format(dms180814[dm_idx])),
                        pd.read_csv('data/CHIME_FRB180814.J0422+73/chime_180814_param_errors_dm{}.csv'.format(dms180814[dm_idx])),
                        how='left', on='name')
bursts180814 = bursts180814.set_index('name')
bursts180814['time_res'] = 0.00098304 # s, Not true for burst 180917

dataframes180814 = []
dmvars_180814 = None
for dm in dms180814:
    frame = pd.merge(pd.read_csv('data/CHIME_FRB180814.J0422+73/chime_180814_fit_params_dm{}.csv'.format(dm)),
                            pd.read_csv('data/CHIME_FRB180814.J0422+73/chime_180814_param_errors_dm{}.csv'.format(dm)),
                            how='left', on='name')
    frame = frame.set_index('name')
    frame['time_res'] = 0.00098304 # s, Not true for burst 180917
    frame['DM']       = dm
    dataframes180814.append(frame)
    if dmvars_180814 is None:
        dmvars_180814 = frame
    else:
        dmvars_180814 = pd.concat([dmvars_180814, frame])
# dmvars_180814.to_csv("universal/data/CHIME_FRB180814.J0422+73/dmvars_180814.csv")
###########
burstsSGR1935 = pd.read_csv('data/SGR1935+2154/SGR1935+2154bursts.csv')
burstsSGR1935 = burstsSGR1935.set_index('name')
##########
# bursts121102.to_csv('../../galacticfrb/data/aggregate/burstsFRB121102.csv', index_label="name")
# bursts180916.to_csv('../../galacticfrb/data/aggregate/burstsFRB180916.csv', index_label="name")
# bursts180814.to_csv('../../galacticfrb/data/aggregate/burstsFRB180814.csv', index_label="name")

# Correct sigmax and sigmay for pixel scale
# bursts180814['sigmax'] = bursts180814[['sigmaxp', 'xscale']].apply(lambda row: abs(row['sigmaxp'])/row['xscale'], axis=1)
# bursts180814['sigmay'] = bursts180814[['sigmayp', 'yscale']].apply(lambda row: abs(row['sigmayp'])/row['yscale'], axis=1)
# bursts180814['sigmax_error'] = bursts180814[['sigmaxp_error', 'xscale']].apply(lambda row: abs(row['sigmaxp_error'])/row['xscale'], axis=1)
# bursts180814['sigmay_error'] = bursts180814[['sigmayp_error', 'yscale']].apply(lambda row: abs(row['sigmayp_error'])/row['yscale'], axis=1)

# Set redshifts
bursts121102['z'] = 0.19273 # (josephy et al. 2019, frbcat.org)
bursts180916['z'] = 0.0337  # (chime 2020b et al.)
bursts180814['z'] = 0.05    # not accurate, z < 0.1 (amiri et al. 2019)
burstsSGR1935['z'] = 0       # ? galactic

for frame in dataframes180916:
    frame['z'] = 0.337
for frame in dataframes180814:
    frame['z'] = 0.05

tauwerror_expr = lambda r: 1e3*r['time_res']*np.sqrt(r['max_sigma']**6*r['min_sigma_error']**2*np.cos(r['angle']-np.pi/2)**4 + r['angle_error']**2*r['max_sigma']**2*r['min_sigma']**2*(-r['max_sigma']**2 + r['min_sigma']**2)**2*np.cos(r['angle']-np.pi/2)**2*np.sin(r['angle']-np.pi/2)**2 + r['max_sigma_error']**2*r['min_sigma']**6*np.sin(r['angle']-np.pi/2)**4)/(r['max_sigma']**2*np.cos(r['angle']-np.pi/2)**2 + r['min_sigma']**2*np.sin(r['angle']-np.pi/2)**2)**1.5
# martin's expr
# tauwerror_expr = lambda r: 1e3*r['time_res']*np.sqrt((r['angle_error']**2*r['max_sigma']**2*r['min_sigma']**2*(-2*r['max_sigma']**2*np.cos(r['angle']-np.pi/2)*np.sin(r['angle']-np.pi/2) + 2*r['min_sigma']**2*np.cos(r['angle']-np.pi/2)*np.sin(r['angle']-np.pi/2))**2)/(4.*(r['max_sigma']**2*np.cos(r['angle']-np.pi/2)**2 + r['min_sigma']**2*np.sin(r['angle']-np.pi/2)**2)**3) + r['min_sigma_error']**2*(-((r['max_sigma']*r['min_sigma']**2*np.sin(r['angle']-np.pi/2)**2)/(r['max_sigma']**2*np.cos(r['angle']-np.pi/2)**2 + r['min_sigma']**2*np.sin(r['angle']-np.pi/2)**2)**1.5) + r['max_sigma']/np.sqrt(r['max_sigma']**2*np.cos(r['angle']-np.pi/2)**2 + r['min_sigma']**2*np.sin(r['angle']-np.pi/2)**2))**2 + r['max_sigma_error']**2*(-((r['max_sigma']**2*r['min_sigma']*np.cos(r['angle']-np.pi/2)**2)/(r['max_sigma']**2*np.cos(r['angle']-np.pi/2)**2 + r['min_sigma']**2*np.sin(r['angle']-np.pi/2)**2)**1.5) + r['min_sigma']/np.sqrt(r['max_sigma']**2*np.cos(r['angle']-np.pi/2)**2 + r['min_sigma']**2*np.sin(r['angle']-np.pi/2)**2))**2)

def computeModelDetails(frame):
    frame['drift_abs'] = -1*(frame['drift (mhz/ms)'])
    frame['drift_over_nuobs'] = frame[['drift_abs','center_f']].apply(lambda row: row['drift_abs'] / row['center_f'], axis=1)
    frame['min_sigma'] = frame[['sigmax','sigmay']].apply(lambda row: min(abs(row['sigmax']), abs(row['sigmay'])), axis=1)
    frame['max_sigma'] = frame[['sigmax','sigmay']].apply(lambda row: max(abs(row['sigmax']), abs(row['sigmay'])), axis=1)
    # the following two lines assume that if sigmax > sigmay, then sigmax_error > sigmay_error, which is true (so far) for this dataset
    frame['min_sigma_error'] = frame[['sigmax_error','sigmay_error']].apply(lambda row: min(row['sigmax_error'], row['sigmay_error']), axis=1)
    frame['max_sigma_error'] = frame[['sigmax_error','sigmay_error']].apply(lambda row: max(row['sigmax_error'], row['sigmay_error']), axis=1)

    frame['sigma_t']   = frame[['min_sigma','time_res']].apply(lambda row: row['min_sigma']*row['time_res'], axis=1)

    frame['tau_w'] = frame[['time_res', 'min_sigma', 'max_sigma', 'angle']].apply(
        lambda r: r['time_res']*r['min_sigma']*r['max_sigma'] / np.sqrt( np.abs((np.sin(r['angle']-np.pi/2)*r['min_sigma'])**2 + (np.cos(r['angle']-np.pi/2)*r['max_sigma'])**2 )), 
        axis=1
    )

    # this error is in ms
    frame['tau_w_error'] = frame[['time_res', 'min_sigma', 'max_sigma', 'min_sigma_error', 'max_sigma_error', 'angle', 'angle_error']].apply(
        tauwerror_expr,
        axis=1
    )

    frame['sigma_t_ms'] = frame['sigma_t']*1e3
    frame['tau_w_ms'] = frame['tau_w']*1e3

    ## Redshift corrections
    frame['drift_z'] = frame[['drift_over_nuobs', 'z']].apply(lambda row: row['drift_over_nuobs']*(1+row['z']), axis=1)
    frame['tau_w_ms_z'] = frame[['tau_w_ms', 'z']].apply(lambda row: row['tau_w_ms']/(1+row['z']), axis=1)

    return frame

burstsSGR1935 = computeModelDetails(burstsSGR1935)
burstsSGR1935
```
